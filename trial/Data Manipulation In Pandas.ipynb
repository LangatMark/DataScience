# series
#series is similar to numpy array. Only that a series has axis labels i.e can be indexed by a label instead of a numeric location like a numpy array. it does not need to hold numeric data and can have any arbitrary object

```python
#create a series from a list, array, and dictionary
label=['a','b','c']#list
my_list=[10,20,10]#list
arr=np.array([10,20,30])#array
d={'a':10,'b':20,'c':30}#dictionary
```

```python
#create a series using a list
pd.Series(my_list)
```

```python
#create a series with labels as axis
pd.Series(data=my_list,index=label)
```

```python
#Create a series from numpy arrays
pd.Series(arr,label)
```

```python
#Create a series from a dictionary
pd.Series(d)
```

```python
pd.Series(arr)
```

```python
#lambda function in mapping...i.e mapping is transforming data as per given condition eg squares
sequences=[10,2,8,7,5,4,3,11,0]
result=map(lambda x: x*x,sequences)
print(list(result))
```

```python

dates_end=pd.date_range(start=start_date,periods=10,freq='T')#T represents minutes
dates_end
```

```python
#create a random data frame with end and start dates above and randomm target value either 0 and 1
import random
randomlist=[]
for i in range(10):
    randomlist.append(random.randint(0,1))
    
#dataframe
df4=pd.DataFrame()
df4['start_date']=dates_start
df4['end_date']=dates_end
df4['Target']=randomlist
df4.head()
```

```python
#extract day from start date
df4['Day']=df4['start_date'].dt.day
df4.head()
```

```python
df4['Month']=df4['start_date'].dt.month
df4.head()
```

```python
#extract year
#hour
#minute
#second
#weekday
#week
#month-year
```

```python
df4['year']=df4['start_date'].dt.year
df4.head()
```

```python
df4['hour']=df4['start_date'].dt.hour
df4.head()
```

```python
df4['minute']=df4['start_date'].dt.minute
df4.head()
```

```python
df4['second']=df4['start_date'].dt.second
df4.head()
```

```python
df4['weekday']=df4['start_date'].dt.weekday
df4.head()
```

```python
df4['week']=df4['start_date'].dt.isocalendar().week
df4.head()
```

```python
#calculating duration
df4['Duration']=df4['end_date']-df4['start_date']
df4.head()
```

```python
#calculating duration in days/hrs/mins
#days
df4['Duration_days']=df4['Duration']/timedelta(days=1)#days
df4['Duration_mins']=df4['Duration']/timedelta(minutes=1)#minutes
df4['Duration_sec']=df4['Duration']/timedelta(seconds=1)#seconds
df4.head()
```
